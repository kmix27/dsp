```python
%matplotlib inline
import nsfg
import thinkstats2
import thinkplot
import random


sample = [random.random() for x in range(0,1000)]


pmf = thinkstats2.Pmf(sample, label='pmf, 1000 random numbers')
cdf = thinkstats2.Cdf(sample, label='cdf, 1000 random numbers')

thinkplot.PrePlot(2)
thinkplot.Pmf(pmf, linewidth=.04)
thinkplot.Show()
thinkplot.Cdf(cdf)
thinkplot.Show(loc = 'lower right')


```
